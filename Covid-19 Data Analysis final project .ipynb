# **Confirmed Cases = Bar Chart**

```python
top10_cases = country_cases.sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top10_cases.index, top10_cases.values)

plt.title("Top 10 Countries by Confirmed COVID-19 Cases")
plt.xlabel("Country")
plt.ylabel("Confirmed cases")
plt.xticks(rotation=45)
plt.show()
```

# **Bar Chart (Confirmed Cases)**

1.The bar chart compares the top 10 countries with the highest confirmed COVID-19 cases.

2.The United States recorded the highest number of confirmed cases among all countries.

3.India ranked second, followed by France, Germany, and Brazil.

4.The chart clearly shows a significant difference between the US and the remaining countries.

5.This visualization makes it easy to compare confirmed cases across the most affected countries.

# **Horizontal Bar Chart**

```python
top10_deaths = country_deaths.sort_values().tail(10)

plt.figure(figsize=(12,6))

plt.barh(top10_deaths.index, top10_deaths.values)

plt.title("Top 10 Countries by Deaths")
plt.xlabel("Deaths")
plt.ylabel("Country")

plt.show()
```

# **Horizontal Bar Chart (Deaths)**

1.The horizontal bar chart displays the top 10 countries with the highest COVID-19 death counts.

2.The United States reported the highest number of deaths during the pandemic.

3.Brazil and India also recorded a large number of deaths compared to other countries.

4.Countries in the lower ranks have noticeably fewer deaths than the leading countries.

5.The chart provides a clear comparison of death counts among the most affected countries.

# **Pie Chart confirmed-cases**

```python
top5 = country_cases.sort_values(ascending=False).head(5)

plt.figure(figsize=(8,8))

explode = (0.1,0,0,0,0)

plt.pie(top5.values,
        labels=top5.index,
        autopct='%1.1f%%',
        explode=explode,
        shadow=True,
        startangle=90)

plt.title("Top 5 Countries by Confirmed Cases")

plt.show()
```

# **Scatter plot = confirmed case**

```python
import matplotlib.pyplot as plt


latest_date = confirmed_df.columns[-1]
country_cases = confirmed_df.groupby('Country')[latest_date].sum()
top10 = country_cases.sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
plt.scatter(
    top10.index,
    top10.values,
    s=150,
    marker='o'
)

plt.title("Top 10 Countries by Confirmed COVID-19 Cases")
plt.xlabel("Country")
plt.ylabel("Confirmed Cases")

plt.xticks(rotation=45)

plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

# **Scatter Plot (Confirmed Cases)**
1.The scatter plot displays confirmed COVID-19 cases for the top 10 countries.

2.Each point represents the total confirmed cases of one country.

3.The United States appears as a clear outlier with the highest number of confirmed cases.

4.Most other countries are grouped below the US with comparatively lower case counts.

5.The scatter plot effectively highlights variations and differences among countries

# **COVID-19 Dashboard (Confirmed Cases)**

```python
import matplotlib.pyplot as plt


latest_date = confirmed_df.columns[-1]
country_cases = confirmed_df.groupby('Country')[latest_date].sum()
top10 = country_cases.sort_values(ascending=False).head(10)
top5 = country_cases.sort_values(ascending=False).head(5)
plt.figure(figsize=(18,10))

# 1. Vertical Bar Chart

plt.subplot(2,3,1)
plt.bar(top10.index, top10.values)
plt.title("Top 10 Confirmed Cases")
plt.xticks(rotation=45)

# 2. Scatter Plot

plt.subplot(2,3,2)
plt.scatter(top10.index, top10.values, s=120, marker='o')
plt.title("Scatter Plot")
plt.xticks(rotation=45)

# 3. Line Chart

plt.subplot(2,3,3)
plt.plot(top10.index, top10.values, marker='o')
plt.title("Line Chart")
plt.xticks(rotation=45)

# 4. Horizontal Bar Chart

plt.subplot(2,3,4)
plt.barh(top10.index, top10.values)
plt.title("Horizontal Bar")

# 5. Pie Chart

plt.subplot(2,3,5)
explode = (0.1,0,0,0,0)

plt.pie(
    top5.values,
    labels=top5.index,
    autopct='%1.1f%%',
    explode=explode,
    shadow=True,
    startangle=90
)

plt.title("Top 5 Confirmed Cases")
plt.suptitle("COVID-19 Confirmed Cases Dashboard", fontsize=18)
plt.tight_layout()

plt.show()
```

# **Dashboard (Combined Visualization)**

1.The dashboard combines multiple visualizations into a single view for better analysis.

2.It allows quick comparison of confirmed cases and deaths using different chart types.

3.The dashboard highlights the countries most affected by the COVID-19 pandemic.

4.Different visualizations provide insights into rankings, proportions, and distributions.

5.Overall, the dashboard offers a comprehensive overview of the COVID-19 dataset, making the analysis easier to understand.

# **COVID-19 Dashboard (Deaths Cases)**

```python
import matplotlib.pyplot as plt


latest_date = deaths_df.columns[-1]
country_deaths = deaths_df.groupby('Country')[latest_date].sum()
top10 = country_deaths.sort_values(ascending=False).head(10)
top5 = country_deaths.sort_values(ascending=False).head(5)
plt.figure(figsize=(18,10))

# 1. Vertical Bar Chart

plt.subplot(2,3,1)
plt.bar(top10.index, top10.values)

plt.title("Top 10 Death Cases")
plt.xlabel("Country")
plt.ylabel("Deaths")
plt.xticks(rotation=45)

# 2. Scatter Plot

plt.subplot(2,3,2)
plt.scatter(
    top10.index,
    top10.values,
    s=120,
    marker='o'
)

plt.title("Scatter Plot")
plt.xlabel("Country")
plt.ylabel("Deaths")
plt.xticks(rotation=45)

# 3. Line Chart

plt.subplot(2,3,3)
plt.plot(
    top10.index,
    top10.values,
    marker='o',
    linewidth=2
)

plt.title("Line Chart")
plt.xlabel("Country")
plt.ylabel("Deaths")
plt.xticks(rotation=45)

# 4. Horizontal Bar Chart

plt.subplot(2,3,4)
plt.barh(top10.index, top10.values)
plt.title("Horizontal Bar")
plt.xlabel("Deaths")
plt.ylabel("Country")

# 5. Pie Chart

plt.subplot(2,3,5)
explode = (0.1,0,0,0,0)
plt.pie(
    top5.values,
    labels=top5.index,
    autopct='%1.1f%%',
    explode=explode,
    shadow=True,
    startangle=90
)

plt.title("Top 5 Death Cases")



plt.suptitle(
    "COVID-19 Death Cases Dashboard",
    fontsize=18,
    fontweight='bold'
)

plt.tight_layout()

plt.show()
```

# **Dashboard (Combined Visualization - Death Cases)**

1.The dashboard combines multiple visualizations to provide a comprehensive analysis of COVID-19 death cases.

2.It enables quick comparison of death counts across the most affected countries using different chart types.

3.The dashboard highlights the countries with the highest COVID-19 fatalities, making trends easy to identify.

4.Each visualization presents a different perspective, including country rankings, proportions, distributions, and comparisons.

5.Overall, the dashboard provides a clear and effective summary of global COVID-19 death data, helping users better understand the pandemic's impact.

<h2 align="center">🌍 About COVID-19</h2>

---



COVID-19 (Coronavirus Disease 2019) is a highly infectious disease caused by the SARS-CoV-2 virus. The first known cases were reported in Wuhan, China, in late 2019. Due to its rapid spread across countries, the World Health Organization (WHO) declared COVID-19 a global pandemic on March 11, 2020.

The pandemic affected millions of people worldwide, leading to widespread illness, hospitalizations, and a significant number of deaths. Governments implemented measures such as lockdowns, travel restrictions, social distancing, and vaccination programs to reduce the spread of the virus.

Countries such as the United States, India, Brazil, France, Germany, and many others experienced a large number of confirmed cases and deaths during different waves of the pandemic. The impact of COVID-19 was not limited to health; it also affected economies, education, businesses, and daily life across the world.

This project analyzes global COVID-19 confirmed and death cases using the Johns Hopkins University dataset. Through data cleaning, exploratory data analysis (EDA), and visualization, the project aims to identify trends, compare the most affected countries, and provide meaningful insights into the global impact of the pandemic.
